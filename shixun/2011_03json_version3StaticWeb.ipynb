```python
ini_s = etree.HTML(ini_html)
ini_content = ini_s.xpath('//*[@id="wrapper"]/div[1]/p[3]/a//@href')
#print(ini_content)
```

```python

queueYear = Queue()
queueLast = Queue()
```

```python
#print('0')
for i in range(1,len(ini_content)):
    print(i,ini_content[i])
    queueYear.put(ini_content[i])
    #print('0')
```

```python
while not queueYear.empty():
    
    #nexturl是不同年份，全部省份，全部文理分科的url
    nexturl = queueYear.get()
    
    next_r = requests.get(nexturl)
    next_html = next_r.text
    next_s = etree.HTML(next_html)
    #获得每个年份的末页
    next_content = next_s.xpath('//*[@id="wrapper"]/div[4]/ul/li[4]/a//@href')
    print(next_content)
    
    if next_content:
        #print(next_content[0])
        queueLast.put(next_content[0])
    
print(queueLast)
```
