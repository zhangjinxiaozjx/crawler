```python
import requests
from bs4 import BeautifulSoup
from lxml import etree
#import pymysql
import json
#import urllib3.request
from collections import OrderedDict
#import httplib
from queue import Queue
import re
```

```python
ini_url = "http://college.gaokao.com/areapoint/p1"
url_model= "http://college.gaokao.com/areapoint/p"
pagenum = 190
URL_JFK = 'http://211.87.230.23:8080'
```

```python
for i in range(1,pagenum+1):
    print("page:",i)
    url = url_model+str(i)
    print(url)
    r = requests.get(url)
    html = r.text
    
    s = etree.HTML(html)
    
    xpath0 = '//*[@id="wrapper"]/div[4]/table/tr'
    
    yearPath = '/td[1]//text()'
    provincePath = '/td[2]//text()'
    subjectPath = '/td[3]//text()'
    batchPath = '/td[4]//text()'
    scorePath = '/td[5]//text()'
    
    #//*[@id="wrapper"]/div[4]/table/tr,/td[1]//text(),/td[2]//text(),/td[3]//text(),/td[4]//text(),/td[5]//text()
    
    #testPath = xpath0+'['+str(2)+']'+yearPath
    #content = s.xpath(testPath)
    #print(content)
    
    
    count_xpath = xpath0+yearPath
    for_count = s.xpath(count_xpath)
    count = len(for_count)
    
    ret = []
    for j in range(count):
        lyear = s.xpath(xpath0+'['+str(j+2)+']'+yearPath)
        if len(lyear)!=0:
            year = lyear[0]
        else:
            year = None
            
        lprovince = s.xpath(xpath0+'['+str(j+2)+']'+provincePath)
        if len(lprovince)!=0:
            province = lprovince[0]
        else:
            province = None
            
        lsubject = s.xpath(xpath0+'['+str(j+2)+']'+subjectPath)
        if len(lsubject)!=0:
            subject = lsubject[0]
        else:
            subject = None
        
        lbatch = s.xpath(xpath0+'['+str(j+2)+']'+batchPath)
        if len(lbatch)!=0:
            batch = lbatch[0]
        else:
            batch = None
        
        
        lscore = s.xpath(xpath0+'['+str(j+2)+']'+scorePath)
        if len(lscore)!=0:
            score = lscore[0]
        else:
            score = None
            
            
        
        
        item = OrderedDict()
        item.clear()
        
        item['province'] = province
        item['year'] = year
        item['category'] = subject
        item['batch'] = batch
        item['grade'] = score
        
        ret.append(item)
        
        #json_str = json.dumps(item,indent=4,ensure_ascii=False)
    data = {'pcutoffs':ret}    
    json_str = json.dumps(data,indent=4,ensure_ascii=False)
    #print(json_str)
    
    headers = {'Content-Type':'Application/json'}
    request = requests.post(URL_JFK+'/addPcutoff',data = json_str.encode('utf-8'),headers=headers)
    r = request.json()
    print("return:",r)
    
    
    #print(content)
    
    
    
    
    
    
```
